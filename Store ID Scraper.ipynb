```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from time import sleep
from selenium.common.exceptions import NoSuchElementException
from datetime import datetime
from selenium.webdriver.chrome.options import Options
```

```python
path = r"C:\Users\Sanchit\Downloads\chromedriver.exe"
mobile_emulation = {
   "deviceMetrics": { "width": 360, "height": 640, "pixelRatio": 3.0 },
   "userAgent": "Mozilla/5.0 (Linux; Android 4.2.1; en-us; Nexus 5 Build/JOP40D) AppleWebKit/535.19 (KHTML, like Gecko) Chrome/18.0.1025.166 Mobile Safari/535.19" }

options = Options()
options.add_experimental_option("mobileEmulation", mobile_emulation)
```

```python
df = pd.read_excel('Chennai, Pune, Kolkata, Gurgaon, Noida, Chandigarh, Lucknow, Jaipur, Vizag & Pune, Ahemdabad, Faridabad.xlsx')
```

```python
df
```

```python
storeid = []
```

```python
i = 0
```

```python
while i<len(df):
    browser = webdriver.Chrome(executable_path=path, chrome_options = options)
    browser.get('https://www.swiggy.com/swiggy-instamart')
    sleep(1)
    browser.find_element_by_xpath('//*[@id="root"]/div/main/section[1]/div/div/div[2]/a/button/div/span').click()
    sleep(1)
    browser.find_element_by_xpath('//*[@id="root"]/div/div[1]/button').click()
    sleep(2)
    browser.find_element_by_xpath('//*[@id="root"]/div/div[1]/div[1]/div/div[1]/label/input').send_keys(str(df.loc[i,"Pincode"]))
    sleep(2)
    try:
        browser.find_element_by_xpath('//*[@id="root"]/div/div[1]/div[2]/div/div/button[1]/span/span[2]').click()
        sleep(2)
        browser.find_element_by_xpath('//*[@id="root"]/div/div/div[3]/div/div[2]/div/div[2]/div[2]/button/div/span[1]').click()
        sleep(2)
        browser.find_element_by_xpath('//*[@id="root"]/div/div/div[1]/div/form/div/input').send_keys('yogabars')
        sleep(2)
        browser.find_element_by_xpath('//*[@id="root"]/div/div/div[2]/div[2]/div/div/div[1]/div/div[2]').click()
        src = browser.current_url
        fin = src.split("=")
        storeid.append(fin[1])
    except NoSuchElementException:
        storeid.append("Location Inservicable")
    
    browser.close()
    i=i+1
```

```python
storeid
```

```python
df["Store ID"] = storeid
```

```python
df.to_excel('rest of the cities Pincode vs Store ID.xlsx')
```
